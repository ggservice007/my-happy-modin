## small records

```python
%%time
import my_happy_modin
import my_happy_modin.my_happy_pandas as pd

lst_json_a = []
lst_json_a.append({
    'name': 'gg008',
    'age': 36
})

lst_json_a.append({
    'name': 'gg009',
    'age': 37
})


lst_json_b = []
lst_json_b.append({
    'name': 'gg008',
    'age': 36
})

lst_json_b.append({
    'name': 'gg010',
    'age': 41
})


df_a = pd.DataFrame.from_records(lst_json_a)

df_b = pd.DataFrame.from_records(lst_json_b)

df_comparison = df_a.merge(df_b, indicator=True, how='outer')

df_comparison
```
